# Vector space
A vector space refers to any set of objects for which addition and scalar multiplication are defined. The operations of vector addition and scalar multiplication must satisfy certain requirements, called vector axioms which are:
- Additive inverse: $\mathbf{v} + (-\mathbf{v}) = \mathbf{0}$
- Associativity: $(\mathbf{v} + \mathbf{w}) + \mathbf{u} = \mathbf{v} + (\mathbf{w} + \mathbf{u})$
- Commutativity: $\mathbf{v} + \mathbf{w} = \mathbf{w} + \mathbf{v}$
- Additive identity: $\mathbf{v} + \mathbf{0} = \mathbf{v}$
- Multiplicative inverse: $\mathbf{v}1 = \mathbf{v}$
- Distributive: $(\mathbf{\alpha + \beta})(\mathbf{v} + \mathbf{w}) = 
\alpha\mathbf{v} + \alpha\mathbf{w} + \beta\mathbf{v} + \beta\mathbf{w}$

# Vector Subspace (Geometric)
A subspace is the set of all points that you can reach by stretching and combining a collection of vectors
# Vector subspace (Algebraic)
A subspace is the set of all points that can be created by all possible linear combinations of vectors. It satisfiers the following conditions:
- Closed under addition and scalar multiplication
- Contains the zeros vector $\mathbf{0}$

# Subset
A subset in contrast to subspace, is a region in space that can have boundaries instead of extending to infinity and it need not include the origin.

# Span (Geometry)
The set of all possible vectors you can reach with linear combinations of a given pair of vectors is called the span of those two vectors.
# Span (Algebraic)
### $span({\mathbf{v}_1,...,\mathbf{v}_n}) = \{\alpha\mathbf{v}_1 + ... + \alpha\mathbf{v}_n, \alpha \in \mathbb{R}\}$

# Linear independence
A set of vectors is independent if the dimensionality of the subspace spanned by that set of vectors is equal to the number of vectors in that set. Algebraically, a set of vectors is dependent if at least one vector in the set can be expressed as a linear weighted combination of the other vectors in that set.
## Example
### $\{\mathbf{w}_1, \mathbf{w}_2\} = \{\begin{bmatrix} 1 \\ 2 \\ 3 \end{bmatrix}, \begin{bmatrix} 2 \\ 4 \\ 6 \end{bmatrix}\}$ &ensp; $\mathbf{w}_2 = 2\mathbf{w}_1$ 
### $\{\mathbf{v}_1, \mathbf{v}_2, \mathbf{v}_3\} = \{\begin{bmatrix} 0 \\ 2 \\ 5 \end{bmatrix}, \begin{bmatrix} -27 \\ 5 \\ -37 \end{bmatrix}, \begin{bmatrix} 3 \\ 1 \\ 8 \end{bmatrix}\}$ &ensp; $\mathbf{v}_2 = 7\mathbf{v}_1 - 9\mathbf{v}_3$
Both examples above show dependent sets
### $\{\mathbf{w}_1, \mathbf{w}_2\} = \{\begin{bmatrix} 1 \\ 2 \\ 3 \end{bmatrix}, \begin{bmatrix} 2 \\ 4 \\ 7 \end{bmatrix}\}$
### $\{\mathbf{v}_1, \mathbf{v}_2, \mathbf{v}_3\} = \{\begin{bmatrix} 0 \\ 2 \\ 5 \end{bmatrix}, \begin{bmatrix} -27 \\ 0 \\ -37 \end{bmatrix}, \begin{bmatrix} 3 \\ 1 \\ 9 \end{bmatrix}\}$
These examples show independent sets

# Basis
A basis is the combination of span and independence. A set of vectors $\{\mathbf{v}_1,\mathbf{v}_2,...,\mathbf{v}_n\}$ forms a basis for some subspace of $\mathbb{R}^N$ if
1. It spans the subspace
2. It is an independent set of vectors.
