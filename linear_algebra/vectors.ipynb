# Geometric Interpretation
A Vector is a line with magnitude (length) and direction in a coordinate system.\
The dimension of the vector correspond to the dimensionality of the coordinate system.\
[2 3] and [3 4 2] are 2D and 3D vectors respectively.\
Vectors can be anywhere on the coordinate system it is not necessary to start from the origin.

```python
import matplotlib.pyplot as plt
import numpy as np

v = np.array([5, -3])
plt.plot([0, v[0]], [0, v[1]])
plt.xlim((-6, 6))
plt.ylim((-6, 6))
plt.title(f"Visualization of {v} on a 2D plane")
plt.grid(True)
plt.show()
```

# Algebraic Interpretation
A vector is an ordered list of numbers, the length of the vector corresponds to the dimensionality of the vector.\
Geometric interpretation of vectors are useful for in 2D and 3D but algebraic perspective allow us to extend to n-dimension.

# Vector Orientation
Vector elements can be stacked vertically (column vector) or laid down horizontally (row vector).\
Vectors are commonly represented in column orientation.\
Column vector: $\begin{bmatrix} 5 \\ 3 \\ 4 \end{bmatrix}$  Row vector: $\begin{bmatrix} 5 & 3 & 4 \end{bmatrix}$

```python
v_col = np.array([[5], [3], [4]])
print("Column Vector:\n", v_col)
```

```python
v_row = np.array([5, 3, 4])
print("Row Vector:\n", v_row)
```

# Vector Operations

## Transpose
Transpose operation converts a column vector to row vector and row vector to column vector.\
The values and the ordering of the elements stays the same.\
$\begin{bmatrix} 5 \\ 3 \\ 4 \end{bmatrix}^T$ = $\begin{bmatrix} 5 & 3 & 4 \end{bmatrix}$\
$\begin{bmatrix} 5 & 3 & 4 \end{bmatrix}^T$ = $\begin{bmatrix} 5 \\ 3 \\ 4 \end{bmatrix}$


## Addition and subtraction
### Geometric
To add two vectors **a** and **b**, place the tail of the vector **b** on top of vector **a**.\
The new vector then goes from the tail of vector **a** to the head of vector **b**.\
\
To subtract two vectors **a** and **b**, you can multiply one of the vector by -1 and them as given above.

```python
v1 = np.array([5, 4])
v2 = np.array([3, -2])
v3 = v1 + v2

fig, ax = plt.subplots(1, 2, figsize=(16, 8))
ax[0].plot([0, v1[0]], [0, v1[1]], label="v1", color="blue")
ax[0].plot([0, v2[0]], [0, v2[1]], label="v2", color="green")
ax[0].plot([0, v3[0]], [0, v3[1]], label="v1 + v2", color="red")
ax[0].set_xlim((-10, 10))
ax[0].set_ylim((-10, 10))
ax[0].legend()

ax[1].quiver(0, 0, v1[0], v1[1], angles="xy", scale_units="xy", color="blue", scale=1, label="v1")
ax[1].quiver(v1[0], v1[1], v2[0], v2[1], angles="xy", scale_units="xy", scale=1, label="v2", color="green")
ax[1].quiver(0, 0, v3[0], v3[1], angles="xy", scale_units="xy", scale=1, label="v1 + v2", color="red")
ax[1].set_xlim((-10, 10))
ax[1].set_ylim((-10, 10))
ax[1].legend()

plt.tight_layout()
plt.show()
```

```python
v1 = np.array([5, 4])
v2 = np.array([3, -2])
v3 = v1 - v2

fig, ax = plt.subplots(1, 2, figsize=(16, 8))
ax[0].plot([0, v1[0]], [0, v1[1]], label="v1", color="blue")
ax[0].plot([0, v2[0]], [0, v2[1]], label="v2", color="green")
ax[0].plot([0, v3[0]], [0, v3[1]], label="v1 - v2", color="red")
ax[0].set_xlim((-10, 10))
ax[0].set_ylim((-10, 10))
ax[0].legend()

ax[1].quiver(0, 0, v1[0], v1[1], angles="xy", scale_units="xy", color="blue", scale=1, label="v1")
ax[1].quiver(v1[0], v1[1], -v2[0], -v2[1], angles="xy", scale_units="xy", scale=1, label="v2", color="green")
ax[1].quiver(0, 0, v3[0], v3[1], angles="xy", scale_units="xy", scale=1, label="v1 - v2", color="red")
ax[1].set_xlim((-10, 10))
ax[1].set_ylim((-10, 10))
ax[1].legend()

plt.title("Vector subtraction")
plt.tight_layout()
plt.show()
```

### Algebraic
Vector addition and subtraction are element-wise addition and subtraction of the two vectors.\
$\begin{bmatrix} 5 & 4 & 3 \end{bmatrix}$ + $\begin{bmatrix} 3 & 5 & 6 \end{bmatrix}$ = $\begin{bmatrix} 8 & 9 & 9 \end{bmatrix}$ \
\
$\begin{bmatrix} 7 \\ 8 \\ 2 \end{bmatrix}$  + $\begin{bmatrix} 5 \\ 4 \\ 1 \end{bmatrix}$ = $\begin{bmatrix} 12 \\ 12 \\ 3 \end{bmatrix}$ \
\
Vector addition and subtraction are valid only if they have the same dimension.

## Vector scalar multiplication
### Geometric
This means multiplying vector by a scalar which essentially scaled the vector or shrink the vector without changing its angle
### Algebraic
It is done element-wise that is multiplying each element of a vector by the scalar

```python
v = np.array([5, 4, 2])
v_scaled = 3 * v

print("v:", v)
print(f"3 * {v}: {v_scaled}")
```
