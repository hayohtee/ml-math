# Function
A function is a relation between inputs and outputs, where each input is related to exactly one output. It is often represented as a lowercase italicized letter $f(x)$ which reads f of x. Here f is the function and x is the input to the function.

## Relation
A relation is a set of ordered pairs that follows some rule. It is a rule that shows how pairs of number relate to each other.

## Variable
The x in $f(x)$ also called an input variable can be a:
- A placeholder for any valid input number.
- A point on a grid to evaluate the function.

In the function $y = f(x)$, y is the output variable (also called dependent variable) and x is the input variable (also called independent variable).

## Vertical and Horizontal line test
These are visual test to help identify functions and their properties.

## Vertical line test
It is a graphical tool for identifying a function. The idea is that you imagine a vertical line at any given x-axis, if that line passes through the graph of the function more than once, then the graph is not defined by the function.

## Horizontal line test
It is used to determine whether the function has multiple mappings between input and output. You imagine a horizontal line on the graph, if the line passes through more than one point then it is not possible to determine a unique value of x for a given y. Note that failing a horizontal line test does not mean that the graph is not a function, instead the horizontal line test help categorize function behaviors.

## Domain
The domain of a function is the set of all possible input values of x.

## Codomain
The codomain is the set of all possible outputs of a function, even if the function definition cannot lead to all those values.

## Range
The set of all possible output values y, which is a subset of the codomain.

## Injective (one-to-one)
A function is injective if each output value has at most one input value. Formally, function $f$ is injective if $f(x_1) = f(x_2)$ is true only if $x_1 = x_2$ for all $x_1, x_2$ in the domain of $f$.

## Surjective (onto)
A function is surjective if every value in the codomain has at least one corresponding input value $x$ in the domain. The function $y = f(x)$ is surjective if for every $y \in R$, there exists an $x \in R$ such that $f(x) = y$

## Bijective (invertible)
A function is bijective when it is both injective and surjective. A function $f$ is bijective if for every y in the codomain, there exists a unique x in the domain such that $f(x) = y$.

# Functions in numpy and sympy

```python
import sympy as sym

pi = sym.symbols("pi")
print("From sympy:")
display(pi)
display(pi + 3)
```

```python
import numpy as np

print("From numpy:")
display(np.pi)
display(np.pi + 3)
```

```python
from sympy.abc import x

expr = 2 * x + 3 * x ** 2
display(expr)
display(expr.subs("x", 1))
```

```python
fx = sym.lambdify(x, expr)
fx(1)
```

```python
fx(np.array([0, 1, 2]))
```

## 1
Recall that the formula for a circle is $r^2 = x^2 + y^2$. Derive math functions $f_p(x)$ and $f_n(x)$ for the positive (upper-half) and negative (lower-half) halves of the circle. Plot the two functions.

```python
import matplotlib.pyplot as plt

r = 2
x = np.linspace(-r, r, 50)
y_pos = np.sqrt(r ** 2 - x ** 2)
y_neg = -y_pos.copy()

plt.figure(figsize=(10, 8))
plt.scatter(x, y_pos, label="$f_p(x)$")
plt.scatter(x, y_neg, label="$f_n(x)$")
plt.legend()
plt.show()
```

## 2
$\sqrt{2}$ can be approximated using the “Babylon method” of estimating roots. It is a simple algorithm and works surprisingly well. Implement the following algorithm in Python:

### $a_n = \frac{1}{2}\big{(} a_{n-1} + \frac{2}{a{n-1}}\big{)}$

where $a_n$ is the $n^{th}$ estimate for $\sqrt{2}$, and the starting value $a_0$ is any number that you think might be reasonable starting guess.

Run this algorithm in a for-loop for five iterations, starting with $a_0 = 1$. Print a table of the iteration number in the first column, the estimate in the second column, and the difference from np.sqrt(2) in the second column.

```python
a_0 = 1.0
a_n: list[float] = [a_0]
sqrt_2 = np.sqrt(2)

print(f"{'n':^5}::{'Approx.':^13}::{'diff':^10}")
print("." * 35)

for n in range(1, 6):
    a_prev = a_n[-1]
    result = 0.5 * (a_prev + (2 / a_prev))
    a_n.append(result)
    print(f"{n:^5}::{result:^13.6f}::{(result - sqrt_2):>10.6f}")
```

## 3
Implement and visualize the following functions in Python, using lambda expressions and numpy. Plot using an x-axis domain of −2 to 4, but note the domain restrictions in functions b and k. Label each function as being injective (but not surjective), surjective (but not injective), bijective, or neither.

```python
a = lambda x : np.max([0, x])
b = lambda x: np.sqrt(x)
c = lambda x: 2*np.pow(x, 2) + 3*np.pow(x, 2)
d = lambda x: np.exp(x)
e = lambda x: np.float64((2*x*(3 * x + 1) *(4*x + 1)))
f = lambda x: x + np.sin(x)
g = lambda x: np.float64(x)
h = lambda x: np.float64(2*x + 1)
i = lambda x: np.cos(x)
j = lambda x: np.tanh(x)
k = lambda x: np.log(x) - 4*np.exp(-(x-1)**2)
l = lambda x: np.float64(1)
```

```python
x = np.linspace(-2, 4, 100)

fig, axes = plt.subplots(3, 4, figsize=(16, 16))

axes[0, 0].plot(x, np.vectorize(a)(x))
axes[0, 0].set_title("a(x)")

axes[0, 1].plot(x, np.vectorize(b)(x))
axes[0, 1].set_title("b(x)")

axes[0, 2].plot(x, np.vectorize(c)(x))
axes[0, 2].set_title("c(x)")

axes[0, 3].plot(x, np.vectorize(d)(x))
axes[0, 3].set_title("d(x)")

axes[1, 0].plot(x, np.vectorize(e)(x))
axes[1, 0].set_title("e(x)")

axes[1, 1].plot(x, np.vectorize(f)(x))
axes[1, 1].set_title("f(x)")

axes[1, 2].plot(x, np.vectorize(g)(x))
axes[1, 2].set_title("g(x)")

axes[1, 3].plot(x, np.vectorize(h)(x))
axes[1, 3].set_title("h(x)")

axes[2, 0].plot(x, np.vectorize(i)(x))
axes[2, 0].set_title("i(x)")

axes[2, 1].plot(x, np.vectorize(j)(x))
axes[2, 1].set_title("j(x)")

axes[2, 2].plot(x, np.vectorize(k)(x))
axes[2, 2].set_title("k(x)")

axes[2, 3].plot(x, np.vectorize(l)(x))
axes[2, 3].set_title("l(x)")

plt.show()
```
